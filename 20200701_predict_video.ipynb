```python
import LoadBatches
from keras.models import load_model
from Models import FCN32, FCN8, SegNet, UNet
import glob
import cv2
import numpy as np
import random
```

```python
video_path = './video/car.flv'

n_classes = 11

key = "fcn8"

method = {
    "fcn32": FCN32.FCN32,
    "fcn8": FCN8.FCN8,
    "segnet": SegNet.SegNet,
    'unet': UNet.UNet}

images_path = "data/dataset1/images_prepped_test/"
segs_path = "data/dataset1/annotations_prepped_test/"

input_height = 320
input_width = 320

colors = [
    (random.randint(
        0, 255), random.randint(
            0, 255), random.randint(
                0, 255)) for _ in range(n_classes)]
```

```python
##########################################################################


def label2color(colors, n_classes, seg):
    seg_color = np.zeros((seg.shape[0], seg.shape[1], 3))
    for c in range(n_classes):
        seg_color[:, :, 0] += ((seg == c) *
                               (colors[c][0])).astype('uint8')
        seg_color[:, :, 1] += ((seg == c) *
                               (colors[c][1])).astype('uint8')
        seg_color[:, :, 2] += ((seg == c) *
                               (colors[c][2])).astype('uint8')
    seg_color = seg_color.astype(np.uint8)
    return seg_color


def getcenteroffset(shape, input_height, input_width):
    short_edge = min(shape[:2])
    xx = int((shape[0] - short_edge) / 2)
    yy = int((shape[1] - short_edge) / 2)
    return xx, yy


images = sorted(
    glob.glob(
        images_path +
        "*.jpg") +
    glob.glob(
        images_path +
        "*.png") +
    glob.glob(
        images_path +
        "*.jpeg"))
segmentations = sorted(glob.glob(segs_path + "*.jpg") +
                       glob.glob(segs_path + "*.png") + glob.glob(segs_path + "*.jpeg"))

```

```python
# m = load_model("output/%s_model.h5" % key)
m = method[key](11, 320, 320)  # 有自定义层时，不能直接加载模型
m.load_weights("output/%s_model.h5" % key)


# for video
videoIn = cv2.VideoCapture(video_path)

# for camera
#videoIn = cv2.VideoCapture(0)

# camera (input) configuration
#frame_in_w = 640
#frame_in_h = 480

#videoIn.set(cv2.CAP_PROP_FRAME_WIDTH, frame_in_w)
#videoIn.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_in_h)
#print("capture device is open: " + str(videoIn.isOpened()))


while True:
    # keyboard input value
    key = cv2.waitKey(1) & 0xFF

    # read video 
    ret, outframe = videoIn.read()
    if (ret):

        #gray = cv2.cvtColor(outframe, cv2.COLOR_BGR2GRAY)
        
        # resize 成 320x320
        xx, yy = getcenteroffset(outframe.shape, input_height, input_width)
        im = outframe[xx:xx + input_height, yy:yy + input_width, :]

        # 使用模型切割
        pr = m.predict(np.expand_dims(im, 0))[0]
        pr = pr.reshape((input_height, input_width, n_classes)).argmax(axis=2)

        
        cv2.imshow("seg_predict_res", label2color(colors, n_classes, pr))
        cv2.imshow('camera or video', im)

    else:
        raise RuntimeError("Error while reading from camera.")

    if key == ord('q'):     # press 'q' to leave while
        break

print('Video Capture end, release camera or video.')
videoIn.release()
cv2.destroyAllWindows()
```
